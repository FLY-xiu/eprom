```python
df = pd.read_pickle('./data/eprom0206.pkl')
df1 = df.loc[(df.mblog_article.str.contains('Cookie'))].reset_index(drop=True)
l_list = os.listdir('article')
# 导入txt文本
for l in l_list:
    t = l.split('.')[0]
    with open('article/{}.txt'.format(t),'r',encoding='utf8') as f:
        b = f.read()
        b = b.replace('\n\n','\n').replace('\n\n','\n').replace('\n','<br><br>')
        b = t+'`'+b
    df1.loc[df1.mblog_article.str.contains(t+'`'+'Cookie Expired'),'mblog_article'] = b
    if t == '爱':
        df.loc[df.quote_article.str.contains(t+'`'+'Cookie Expired'),'quote_article'] = b
# 整理非eprom的文章
for i in df1.loc[df1.mblog_name != 'eprom'].index:
    b = df1.loc[i,'mblog_article']
    df1.loc[i,'mblog_article'] = b.replace('Cookie Expired',df1.loc[i,'mblog'])
# 重置所有文章
for i in df1.index:
    url = df1.loc[i,'mblog_article_url']
    b = df1.loc[i,'mblog_article']
    df.loc[df.mblog_article_url == url,'mblog_article'] = b
    df.loc[df.quote_article_url == url,'quote_article'] = b
for i in df.loc[df.quote_name != 'eprom'].index:
    b = df.loc[i,'quote_article']
    df.loc[i,'quote_article'] = b.replace('Cookie Expired',df.loc[i,'quote'])
df.mblog_article = df.mblog_article.str.replace('\n\n','\n').str.replace('\n\n','\n').str.replace('\n','<br><br>')
df.quote_article = df.quote_article.str.replace('\n\n','\n').str.replace('\n\n','\n').str.replace('\n','<br><br>')
df.loc[(df.quote_article.str.contains('Network Error')),'quote_article'] = ''
df.loc[(df.mblog_article.str.contains('Network Error')),'mblog_article'] = ''
df.loc[(df.quote_qa.str.contains('Network Error')),'quote_qa'] = ''
df.loc[(df.mblog_qa.str.contains('Network Error')),'mblog_qa'] = ''
df.to_pickle('./data/eprom.pkl')
```
